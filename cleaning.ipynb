```python
df_age =  df[['PassengerId','Survived','Pclass','Name','Sex','SibSp','Parch','Ticket','Fare','Embarked','Age']]
```

```python
df_age['Ticket']  = df_age['Ticket'].str.strip()
df_age['Ticket']  = df_age['Ticket'].str.lower()
```

```python
df_age["prefix"] = df_age["Ticket"].str.extract(r"([A-Za-z./]+)")
df_age["ticket_nember"] =  df_age["Ticket"].str.extract(r'(\d+)$')
```

```python
df_age["prefix"] = df_age["prefix"].fillna('unknown')
df_age["ticket_nember"] = df_age["ticket_nember"].fillna(0).astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_age['Name'] = le.fit_transform(df_age['Name'])
df_age['Sex'] = le.fit_transform(df_age['Sex'])
df_age['Ticket'] = le.fit_transform(df_age['Ticket'])
df_age['Embarked'] = le.fit_transform(df_age['Embarked'])
df_age['prefix'] = le.fit_transform(df_age['prefix'])
```

```python
predictors = ['PassengerId','Survived','Pclass','Name','Sex','SibSp','Parch','Ticket','Fare','Embarked']
x_traing =  df_age[df_age['Age'].notnull()][predictors]
y_traing =  df_age[df_age['Age'].notnull()]['Age']
x_testing =  df_age[df_age['Age'].isnull()][predictors]
```

```python
x_testing
```

```python
from sklearn.ensemble import RandomForestRegressor
model =  RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_traing, y_traing)
```

```python
predicted_ages = model.predict(x_testing)
```
